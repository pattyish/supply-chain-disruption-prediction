# Exploratory Analysis

This notebook bootstraps quick EDA: it searches for CSV files in the repo, loads the first one it finds, and prints basic summaries. Update the path or add your dataset under a `data/` folder.

```python
from pathlib import Path
import pandas as pd
import numpy as np

# Find CSV files under the repository root
repo_root = Path.cwd().parent  # notebooks/ -> repo root
csv_files = list(repo_root.rglob('*.csv'))
print('Found', len(csv_files), 'CSV files')
dataset = None
if csv_files:
    path = csv_files[0]
    print('Loading', path)
    dataset = pd.read_csv(path)
    print('Shape:', dataset.shape)
else:
    print('No CSV files detected. Place data in a `data/` folder or update the path.')
```

```python
# Quick EDA if dataset is loaded
if dataset is not None:
    display(dataset.head())
    print('Info:')
    print(dataset.info())
    print('Describe:')
    display(dataset.describe(include='all'))
    # Missing values summary
    miss = dataset.isna().sum()
    print('Missing values per column:')
    print(miss[miss>0])
else:
    print('No dataset loaded yet.')
```
